```python
# Convert date in checkin to array
import json
import pandas as pd

# Open the original data
with open('yelp/yelp_academic_dataset_checkin.json', 'r', encoding='utf-8') as infile:
    new_data = []
    for line in infile:
        try:
            data = json.loads(line)
            if 'date' in data and data['date'] is not None:
                data['date'] = data['date'].split(', ')
            new_data.append(data)
        except json.JSONDecodeError:
            pass

# Write the processed data into a new json file
with open('yelp/yelp_academic_dataset_checkin_new.json', 'w', encoding='utf-8') as outfile:
    for data in new_data:
        json.dump(data, outfile)
        outfile.write('\n')

# Close files
infile.close()
outfile.close()
```

```python
# Add checkin to business
import json
import pandas as pd

checkin_df = pd.read_json('yelp/yelp_academic_dataset_checkin_new.json', lines=True)
business_df = pd.read_json('yelp/yelp_academic_dataset_business.json', lines=True)

result_df = business_df.merge(checkin_df[['business_id', 'date']], on='business_id', how='left')

# Write the processed data into a new json file
result_df.to_json('yelp/yelp_academic_dataset_business_new.json', orient='records', force_ascii=False, lines=True)
```

```python
# Process business
import json
import pandas as pd

# Open the original data
with open('yelp/yelp_academic_dataset_business_new.json', 'r', encoding='utf-8') as infile:
    new_data = []

    for line in infile:
        try:
            data = json.loads(line)
            if 'business_id' in data:
                data['_id'] = data.pop('business_id')
            if 'date' in data:
                data['checkin_date'] = data.pop('date')
            if 'attributes' in data and data['attributes'] is not None:
                attributes = data['attributes']

                for field, value in attributes.items():
                    try:
                        dicti = eval(value)
                        data["attributes"][field] = dicti
                    except json.JSONDecodeError:
                        pass
            if 'categories' in data and data['categories'] is not None:
                categories_list = data['categories'].split(', ')
                data['categories'] = categories_list

            new_data.append(data)
        except json.JSONDecodeError:
            pass

# Write the processed data into a new json file
with open('yelp/yelp_academic_dataset_business_new.json', 'w', encoding='utf-8') as outfile:
    for data in new_data:
        json.dump(data, outfile)
        outfile.write('\n')

# Close files
infile.close()
outfile.close()
```

```python
# Process user
import json
import pandas as pd

# Open the original data
with open('yelp/yelp_academic_dataset_user.json', 'r', encoding='utf-8') as infile:
    new_data = []

    for line in infile:
        try:
            data = json.loads(line)
            if 'user_id' in data:
                data['_id'] = data.pop('user_id')
            if 'friends' in data and data['friends'] is not None:
                data['friends'] = data['friends'].split(', ')
            if 'elite' in data and data['elite'] is not None:
                data['elite'] = data['elite'].split(',')
            new_data.append(data)
        except json.JSONDecodeError:
            pass

# Write the processed data into a new json file
with open('yelp/yelp_academic_dataset_user_new.json', 'w', encoding='utf-8') as outfile:
    for data in new_data:
        json.dump(data, outfile)
        outfile.write('\n')

# Close files
infile.close()
outfile.close()
```
